```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv('eda_data.csv')
```

## choose relevant column

```python
df.columns
```

```python
df_model=df[['avg_salary','Rating','Size','Type of ownership', 'Industry', 'Sector','Revenue','Location','num_comp','age',
       'python_yn',  'spark_yn', 'aws_yn', 'excel_yn','job_simp',
       'seniority', 'desc_len']]
```

```python
df_model
```

# get dummy data

```python
df_dum=pd.get_dummies(df_model)
```

```python
df_dum.shape
```

# Train test split

```python
from sklearn.model_selection import train_test_split
X=df_dum.drop('avg_salary',axis=1)
y=df_dum.avg_salary.values
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.4,random_state=42)
```

```python
X_train.shape
```

# Multiple linear regression

```python
import statsmodels.api as sm
```

```python
X_sm=X=sm.add_constant(X)
```

```python
model=sm.OLS(y,X_sm)
model.fit().summary()
```

```python
from sklearn.linear_model import LinearRegression,Lasso
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
```

```python
lm_model=LinearRegression()
```

```python
s=StandardScaler()
```

```python
 s.fit(X_train)
```

```python
X_train_sclaed=s.transform(X_train)
```

```python
lm_model.fit(X_train,y_train)
```

```python
np.mean(cross_val_score(lm_model,X_train,y_train,scoring='neg_mean_absolute_error',cv=5))
```

# Lasso regression

```python
lm_l=Lasso()
lm_l.fit(X_train,y_train)
```

```python
np.mean(cross_val_score(lm_l,X_train,y_train,scoring='neg_mean_absolute_error',cv=3))
```

```python
alpha=[]
error=[]
```

```python
for i in range(1,100):
    alpha.append(i/100)
    lml=Lasso(alpha=(i/100))
    error.append(np.mean(cross_val_score(lml,X_train,y_train,scoring='neg_mean_absolute_error',cv=3)))
```

```python
plt.plot(alpha,error)
```

```python
err=tuple(zip(alpha,error))
```

```python
df_error=pd.DataFrame(err,columns=['alpha','error'])
```

```python
df_error[df_error.error==max(df_error.error)]
```

# Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf=RandomForestRegressor()
```

```python
np.mean(cross_val_score(rf,X_train,y_train,scoring='neg_mean_absolute_error',cv=3))
```

# Tune with Gridsearchcv

```python
 from sklearn.model_selection import GridSearchCV
```

```python
parameters = {'n_estimators':range(10,300,10), 'criterion':('mse','mae'),'max_features':('auto','sqrt',
                                                                                        'log2')}
```

```python
gs=GridSearchCV(rf,parameters,scoring="neg_mean_absolute_error",cv=3)
```

```python
gs.fit(X_train,y_train)
```

```python
gs.best_score_
```

```python
gs.best_estimator_
```

```python
tpred=lm_model.predict(X_test)
```
