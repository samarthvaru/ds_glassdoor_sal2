#Salary parsing

```python
df=df[df['Salary Estimate'] != '-1']
```

```python
df
```

```python
salary=df['Salary Estimate'].apply(lambda x: x.split('<')[0])
```

```python
salary_new=salary.apply(lambda x: x.split(' ')[1:])
```

```python
fsal=salary_new.apply(lambda x: ','.join(x))
```

```python
sal=fsal.apply(lambda x: x.replace(',',''))
```
