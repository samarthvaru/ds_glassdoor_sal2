```python
import pandas as pd
```

```python
df=pd.read_csv('uncleaned_data.csv')
```

```python
df
```

```python
df=df.drop('Unnamed: 0',axis=1)
```

#Salary parsing

```python
df=df[df['Salary Estimate'] != '-1']
```

```python
df
```

```python
salary=df['Salary Estimate'].apply(lambda x: x.split('<')[0])
```

```python
salary_new=salary.apply(lambda x: x.split(' ')[1:])
```

```python
fsal=salary_new.apply(lambda x: ','.join(x))
```

```python
sal=fsal.apply(lambda x: x.replace(',','').replace('K','').replace('₹','').replace('£',''))
```

```python
sal
```

```python
df['min_salary']=sal.apply(lambda x: x.split('-')[0])
```

```python
df['min_salary']
```

```python
df['max_salary']=sal.apply(lambda x: x.split('-')[1])
```

```python
df
```

```python
df['min_salary'].dropna()
```

```python
df['max_salary'].dropna()
```

```python
df=df.astype({'min_salary':float,'max_salary':float})
```

```python
df.min_salary.dtype
```

```python
int(df['min_salary'][1]+df['max_salary'][1])
```

```python
df['avg_salary']=(df.min_salary+df.max_salary)/2
```

```python
df.head()
```

```python
df['Company Name']=df.apply(lambda x: x['Company Name'] if x['Rating'] <0 else x['Company Name'][:-3] , axis=1 )
```

```python
df.columns
```

```python
df['same_state']=df.apply(lambda x: 1 if x['Location']==x['Headquarters'] else 0,axis=1)
```

```python
df['age']=df.Founded.apply(lambda x: x if x<1 else 2020-x)
```
